```python
#Get the data and clean the data
t=requests.get("http://www.the-numbers.com/movie/records/All-Time-Domestic-Box-Office")
soup=BeautifulSoup(t.text)
soup.prettify()
tgm=soup.find_all("tr")
topGrossingMovies={'Movie Name':[],"Year":[],"Domestic Box Office":[], "International Box Office":[],"Worldwide Box Office": []}
for i in range(1, len(tgm)):
    x= tgm[i].find_all("td")
    topGrossingMovies['Year'].append(int(x[1].get_text().strip()))
    topGrossingMovies['Movie Name'].append(x[2].get_text().strip())    
    topGrossingMovies['Domestic Box Office'].append(int(x[3].get_text().replace('$','').replace(',','').strip()))
    topGrossingMovies['International Box Office'].append(int(x[4].get_text().replace('$','').replace(',','').strip()))
    topGrossingMovies['Worldwide Box Office'].append(int(x[5].get_text().replace('$','').replace(',','').strip()))
topGrossingMovies
```

```python
# Put the data into framework
bdf=pd.DataFrame(topGrossingMovies)
bdf=bdf[['Movie Name','Year','Domestic Box Office','International Box Office','Worldwide Box Office']]
bdf.index=bdf.index +1
bdf
```

```python
bdf.to_csv("TopGrossingMovies.csv")
```
